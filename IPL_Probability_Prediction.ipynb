```python
total_scores=delivery.groupby(['match_id', 'inning']).sum()['total_runs'].reset_index()
total_scores
total_scores=total_scores[total_scores['inning']==1]#original total_scores not harmed
total_scores=total_scores.drop(columns='inning')
total_scores
```

```python
match_df=pd.merge(match,total_scores,on='match_id',how='outer')
match_df
```

```python
match_df['team1'].unique()
```

```python
l=['Pune Warriors','Kochi Tuskers Kerala','Rising Pune Supergiants','Gujarat Lions','Rising Pune Supergiant']
match_df=match_df[~match_df.isin(l)]
match_df
```

```python
replacements={ 'Delhi Daredevils':'Delhi Capitals','Deccan Chargers':'Sunrisers Hyderabad'}
match_df=match_df.replace(replacements)
match_df.head(20)
```

```python
l=['team1','team2','toss_winner']
match_df=match_df.dropna(subset=l)
match_df
```

```python
match_df=match_df[match_df['dl_applied']==0]
match_df
```

```python
match_df=match_df[['match_id','city','winner','total_runs']]
match_df
```

```python
delivery_df=pd.merge(match_df,delivery,on='match_id')
delivery_df
```

```python
delivery_df=delivery_df[delivery_df['inning']==2]
delivery_df
```

```python
delivery_df['current_score']=delivery_df.groupby('match_id').cumsum()['total_runs_y']
delivery_df
```

```python
# balls left
delivery_df['balls_left']=126-(delivery_df['over']*6+delivery_df['ball'])
delivery_df
```

```python
# runs left
delivery_df['runs_left']=delivery_df['total_runs_x']-delivery_df['current_score']+1
delivery_df
```

```python
#wickets remaining
# on every player dismissed add 1 so cum sum is the wickets and 10 minus that would be wickets remaining
# but first player dismissed is to be converted to numbers -- fillna with zero and remaining with 1
delivery_df['player_dismissed']=delivery_df['player_dismissed'].fillna(0)
delivery_df
```

```python
def result(row):
    if row['player_dismissed'] != 0:
        return 1
    else:
        return 0

delivery_df['result_column'] = delivery_df.apply(result, axis=1)
delivery_df
```

```python
delivery_df=delivery_df.drop(columns='player_dismissed')
delivery_df
delivery_df=delivery_df.rename(columns={'result_column':'player_dismissed'})
delivery_df
```

```python
delivery_df['wickets']=delivery_df.groupby('match_id').cumsum()['player_dismissed']
delivery_df.head(50)
```

```python
delivery_df['wickets_left']=10-delivery_df['wickets']
delivery_df=delivery_df.drop(columns='wickets')
delivery_df
```

```python
#result
def result(row):
    if row['winner']==row['batting_team']:
        return 1
    else:
        return 0
delivery_df.apply(result,axis=1)
delivery_df['result']=delivery_df.apply(result,axis=1)
delivery_df
```

```python
#crr
delivery_df['crr']=(delivery_df['current_score']*6)/(120-delivery_df['balls_left'])
delivery_df
```

```python
#rrr
delivery_df['rrr']=(delivery_df['runs_left']*6)/delivery_df['balls_left']
delivery_df
```

```python
final_df=delivery_df[['batting_team','bowling_team','city','total_runs_x','runs_left','balls_left','wickets_left','crr','rrr','result']]
```

```python
import numpy as np
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trf=ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','city'])
]
,remainder='passthrough')
```

```python
# Define a function to create your Keras model
def create_model():
    model = Sequential([
        Dense(50, activation='sigmoid', input_shape=(52,)),
        Dense(20, activation='sigmoid'),
        Dense(20, activation='sigmoid'),
        Dense(10, activation='sigmoid'),
        Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

# Create a KerasClassifier using the create_model function
keras_clf = KerasClassifier(build_fn=create_model, epochs=30, batch_size=50, verbose=0)
```

```python
pipeline = Pipeline([
    ('step1', trf),
    ('model', keras_clf)  # Neural network model
])
```

```python
pipeline.fit(x_train, y_train)
```

```python
y_pred = pipeline.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Test accuracy: {accuracy:.4f}")
```
